1. Load Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Dataset/student/student-mat.csv", sep=";")
```

```python
df.head
```

```python
print(df.columns)
```

2. Explore & Clean Data

Check Missing Values

```python
df.isnull().sum()
```

Remove Duplicates

```python
df = df.drop_duplicates()
```

Inspect Dataset Shape & Data Types

```python
print("Shape of dataset:", df.shape)
print("\nData Types:\n")
print(df.dtypes)
```

3. Analysis Questions

Average Final Grade (G3)

```python
avg_grade = df["G3"].mean()
print("Average Final Grade (G3):", avg_grade)
```

Students Scoring Above 15

```python
above_15 = df[df["G3"] > 15].shape[0]
print("Students Scoring Above 15:", above_15)
```

Study Time vs Performance Correlation

```python
correlation = df["studytime"].corr(df["G3"])
print("Correlation Between Study Time and Performance:", correlation)
```

Gender Performance Comparison

```python
gender_avg = df.groupby("sex")["G3"].mean()
print(gender_avg)
```

4. Visualizations

Histogram of Grades

```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(df["G3"])
plt.title("Histogram of Final Grades")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")
plt.show()
```

Scatter Plot (Study Time vs Grades)

```python
plt.figure()
plt.scatter(df["studytime"], df["G3"])
plt.title("Study Time vs Final Grades")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()
```

Bar Chart (Male vs Female Average Score)

```python
plt.figure()
gender_avg.plot(kind="bar")
plt.title("Male vs Female Average Score")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```
